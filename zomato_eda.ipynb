```python
#Handling missing values
df.isnull().sum()
```

The cuisine column has 9 missing values, as it is a categorical feature we can impute missing values with the most frequent cuisine.

```python
df.Cuisines.unique()
```

```python
#finding mode and imputing in cuisine column
x = df['Cuisines'].mode()[0]
df['Cuisines'] = df['Cuisines'].fillna(x)
```

```python
df.isnull().sum()
```

```python
#checking for duplicate records
df.duplicated().sum()
```

There is no duplicate record present in our dataset

Joining the country dataset with main dataset to perform analysis using Country code

```python
ratings = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

More than 2000 restaurants has not been rated or rated 0.

Nearly 300 restaurants have been rated excellent.

```python
plt.figure(figsize=(12,6))
sns.barplot(x=ratings['Aggregate rating'], y=ratings['Rating Count'], hue=ratings['Rating text'], palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

The restaurants are rated in range of 0 t0 5.

Highest rating given to a restaurant is 4.9 and lowest to 0.




Creating a subset from our dataset for India.

```python
df_in = df[df['Country']=="India"]
```

```python
df_in.head()
```

```python
df_in.shape
```

```python
df_in['Votes'].max()
```

```python
#Top 10 voted restaurant in India
top_voted = df_in[['Restaurant Name', 'Votes']].sort_values(by='Votes', ascending=False).head(10).reset_index(drop=1)
top_voted
```

We can see the top 10 voted restaurant.

Toit is at number 1 with 10934 votes.


```python
sns.barplot(y=top_voted['Restaurant Name'], x=top_voted['Votes'], errorbar=None)
plt.show()
```

```python
#Restaurants with highest aggregate rating
df_in[df_in['Aggregate rating']==4.9][['Restaurant Name', 'City', 'Locality']].reset_index(drop=True)
```

There are close to 20 restaurants which were rated 4.9 in India.

```python
#Top 10 Cities that have max numbers of excellent rated restaurants
city_excellent = df_in[df_in['Rating text'] == "Excellent"].groupby(['City']).size().sort_values(ascending=False)[:10]
city_excellent
```

New Delhi has majority of the restaurants which were rated excellent all over India, followed by Gurgaon and Bangalore.

```python
city_excellent.plot(kind='bar')
plt.title("Top 10 Cities with Excellent rated Restaurants")
plt.show()
```

```python
#Top 10 cuisines served
top_cuisines =  df_in.Cuisines.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
top_cuisines.value_counts()[:10]
```

Approxly 4000 restaurants serve North Indian cuisine, followed by Chinese and Fast Food.

```python
plt.figure(figsize=(10,8))
plt.pie(x=top_cuisines.value_counts()[:10], labels=top_cuisines.value_counts()[:10].index, autopct='%1.2f%%')
plt.title("Distribution of top 10 cuisines served")
plt.show()
```
